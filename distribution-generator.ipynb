```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
from sklearn.model_selection import train_test_split

white_wine = pd.read_csv('https://query.data.world/s/brho7trfnuy4iz2nxraqzsfwvgek5b')

high_quality_i = white_wine.loc[ white_wine['quality'] > 6 ]   #ratings 9, 8, 7 only, 1060 rows
mid_quality_initial_i = white_wine.loc[ white_wine['quality'] < 7 ]
mid_quality_i = mid_quality_initial_i.loc[ mid_quality_initial_i['quality'] > 4 ]   #ratings 5, 6 only, 3655 rows
low_quality_i = white_wine.loc[ white_wine['quality'] < 5 ]   #ratings 3, 4 only, 183 rows

high_quality_i['quality'] = 3
mid_quality_i['quality'] = 2
low_quality_i['quality'] = 1

frames = [high_quality_i, mid_quality_i, low_quality_i]
white_wine_new = pd.concat(frames)

X_initial = white_wine_new.drop(['quality'], axis=1)
X = X_initial[ ['alcohol', 'density', 'total sulfur dioxide', 'free sulfur dioxide', 'volatile acidity', 'fixed acidity', 'chlorides', 'residual sugar', 'citric acid', 'sulphates', 'pH'] ]
y = white_wine_new['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
train_set_quality = np.array(y_train)
X_train['quality'] = train_set_quality

high_quality = X_train.loc[ X_train['quality'] == 3 ]
mid_quality = X_train.loc[ X_train['quality'] == 2 ]
low_quality = X_train.loc[ X_train['quality'] == 1 ]
```

```python
#Alcohol Analysis

#High quality wine
high_quality_alcohol_dist = np.empty(10000)

for i in range(10000):
    random_sample = np.random.choice( high_quality['alcohol'], size=10 )
    high_quality_alcohol_dist[i] = np.mean(random_sample)

#Mid quality wine
mid_quality_alcohol_dist = np.empty(10000)

for i in range(10000):
    random_sample = np.random.choice( mid_quality['alcohol'], size=10 )
    mid_quality_alcohol_dist[i] = np.mean(random_sample)

#Low Quality
low_quality_alcohol_dist = np.empty(10000)

for i in range(10000):
    random_sample = np.random.choice( low_quality['alcohol'], size=10 )
    low_quality_alcohol_dist[i] = np.mean(random_sample)
```
