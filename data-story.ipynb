```python
mean_values = white_wine.groupby('quality').agg(np.mean)
print('Mean value analysis')

plt.title('Chlorides')
plt.scatter(mean_values.index, mean_values['chlorides'])
plt.ylim(0.0, 0.055)
plt.show()

plt.title('Total Sulfur Dioxide')
plt.scatter(mean_values.index, mean_values['total sulfur dioxide'])
plt.ylim(100, 180)
plt.ylabel("not '0' indexed")
plt.show()

plt.title('Density')
plt.scatter(mean_values.index, mean_values['density'])
plt.ylim(0.9900, 1)
plt.ylabel("not '0' indexed")
plt.show()

plt.title('pH')
plt.scatter(mean_values.index,mean_values['pH'])
plt.ylim(3, 3.4)
plt.ylabel("not '0' indexed")
plt.show()

plt.title('Alcohol')
plt.scatter(mean_values.index, mean_values['alcohol'])
plt.ylim(9, 13)
plt.ylabel("not '0' indexed")
plt.show()

plt.title('Residual Sugar')
plt.scatter(mean_values.index, mean_values['residual sugar'])
plt.ylim(3, 8)
plt.ylabel("not '0' indexed")
plt.show()

plt.tight_layout()
```

A lower 'residual sugar' seems to correlate to higher quality wine. A lower 'chloride' content correlates with higher quality wine; the histogram analysis found the same correlation. A 'total sulfur dioxide content' greater than 120 correlates with lower quality wine; histogram analysis found a lower bound correlation in which a 'total sulfur dioxide' content less than 75 correlates with a low quality wine. A density of 0.992 and below correlates to a higher quality wine; the histogram analysis found the same correlation. Higher pH correlates to a higher quality wine; histrogram analysis confirms this. Higher alcohol content correlates to a higher quality wine; histrogram analysis confirms this.
